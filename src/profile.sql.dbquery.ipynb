```sql
CREATE TABLE employees (
    emp_id INT,
    emp_name STRING,
    department STRING,
    salary INT
);

INSERT INTO employees VALUES
(1,'John','IT',60000),
(2,'Alice','HR',45000),
(3,'Bob','IT',70000),
(4,'David','Sales',50000),
(5,'Emma','HR',48000),
(6,'Sam','IT',65000),
(7,'Tom','Sales',52000),
(8,'Lisa','HR',47000);

select department,avg(salary) from employees group by department;
select * from employees where emp_name='Alice'

SELECT *
FROM employees
WHERE YEAR(join_date)=2025;

SELECT *
FROM employees
WHERE join_date >= '2025-01-01'
AND join_date < '2026-01-01';
```
