```python
%sql
CREATE TABLE Employee
(
    EmpID INT PRIMARY KEY,
    EmpName VARCHAR(50),
    DeptID INT
);
CREATE TABLE Department
(
    DeptID INT PRIMARY KEY,
    DeptName VARCHAR(50)
);
```

```python
%sql
CREATE TABLE EmployeeSales
(
    SaleID INT PRIMARY KEY,
    EmpID INT,
    EmpName VARCHAR(50),
    Department VARCHAR(30),
    SaleDate DATE,
    MonthName VARCHAR(10),
    Product VARCHAR(50),
    Amount DECIMAL(10,2),
    Email VARCHAR(100)
);
INSERT INTO EmployeeSales
VALUES
(1,101,'John','IT','2026-01-05','Jan','Laptop',5000,'john@gmail.com'),
(2,101,'John','IT','2026-02-10','Feb','Mouse',800,'john@gmail.com'),
(3,101,'John','IT','2026-03-12','Mar','Keyboard',1200,'john@gmail.com'),

(4,102,'Mary','HR','2026-01-08','Jan','Chair',3000,'mary@gmail.com'),
(5,102,'Mary','HR','2026-02-20','Feb','Desk',4500,'mary@gmail.com'),

(6,103,'David','IT','2026-01-15','Jan','Laptop',7000,'david@gmail.com'),
(7,103,'David','IT','2026-03-18','Mar','Monitor',2500,'david@gmail.com'),

(8,104,'Emma','Finance','2026-02-25','Feb','Printer',6000,'emma@gmail.com'),

(9,105,'Alice','HR','2026-03-10','Mar','Desk',5000,'alice@gmail.com'),

-- Duplicate Record
(10,101,'John','IT','2026-03-12','Mar','Keyboard',1200,'john@gmail.com'),

-- Invalid Email (Data Quality)
(11,106,'Bob','Sales','2026-03-15','Mar','Scanner',4000,NULL);
```

```python
%sql

SELECT * from employeesales qualify ROW_NUMBER()OVER(PARTITION BY Department order by amount desc)=1 ;
```

```python
%sql

SELECT *
FROM
(
    SELECT
        EmpName,
        MonthName,
        Amount
    FROM employeesales
)

PIVOT
(
    SUM(Amount)
    FOR MonthName IN ('Jan','Feb','Mar')
);
```

```python
%sql
WITH CTE AS
(
    SELECT *,
           ROW_NUMBER() OVER
           (
               PARTITION BY EmpID, Product, SaleDate
               ORDER BY SaleID
           ) AS rn
    FROM EmployeeSales
)
SELECT *
FROM CTE
WHERE rn = 1;
```
