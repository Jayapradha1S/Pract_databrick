```python
df.select("name", "salary")
df.select(col("name"), col("salary"))
```

```python
df.where(col("salary") > 50000)
df.filter(col("salary") > 50000)
df.filter("salary > 50000")
```

```python
df.withColumn("bonus", col("salary") * 0.10)
df.withColumn("country", lit("India"))
df.withColumnRenamed("salary", "emp_salary")
df.drop("bonus")
df.toDF("id", "name", "salary", "department")
df.groupBy("department").count()
df.groupBy("department").agg(sum("salary"), avg("salary")
df.groupBy("department").agg(sum("salary").alias("total_salary"), avg("salary").alias("avg_salary"))
df.groupBy("department").agg(sum("salary").alias("total_salary"), avg("salary").alias("avg_salary"))
                                                
```
